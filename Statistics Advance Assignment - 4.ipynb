```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
Ans:The t-test and z-test are both statistical tests used to compare means. However, they differ in their assumptions about the population standard deviation.

T-test:

Assumes that the population standard deviation is unknown.
Uses the sample standard deviation to estimate the population standard deviation.
Appropriate when the sample size is small (typically less than 30).
Z-test:

Assumes that the population standard deviation is known.
Uses the known population standard deviation directly.
Appropriate when the sample size is large (typically greater than 30).
Example Scenario:

Suppose you want to compare the average heights of male and female students in a school.

If you know the population standard deviation of heights for both males and females, you can use a z-test. For instance, if previous studies have established that the population standard deviation for male heights is 3 inches and for female heights is 2.5 inches, you can use a z-test to determine if there is a significant difference in average heights between the two groups.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests.
Ans:One-tailed test:

Hypothesis: The alternative hypothesis specifies a direction (greater than or less than).   
Example: The average IQ of students at this school is greater than 100.
Critical region: Exists in only one tail of the distribution (either the right or left tail).   
Power: Has higher power than a two-tailed test when the effect is in the predicted direction.   
Two-tailed test:

Hypothesis: The alternative hypothesis does not specify a direction.   
Example: The average IQ of students at this school is different from 100.
Critical region: Exists in both tails of the distribution.   
Power: Has lower power than a one-tailed test when the effect is in the predicted direction, but it is more robust to unexpected outcomes.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
Ans:Type 1 error:
Definition: Rejecting a null hypothesis when it is actually true.
Example: A researcher concludes that a new drug is effective in treating a disease, but in reality, it has no effect.
Significance level (α): The probability of committing a Type 1 error. It is set before conducting the test.
Type 2 error:

Definition: Failing to reject a null hypothesis when it is false.
Example: A researcher concludes that a new drug is not effective in treating a disease, but in reality, it is effective.
Power: The probability of correctly rejecting a false null hypothesis. It is related to the sample size and the effect size.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Ans:Confidence interval is a range of values within which we are reasonably confident that the true population parameter lies. It is a statistical tool used to estimate the population parameter based on sample data.   

Components of a confidence interval:

Point estimate: The best estimate of the population parameter from the sample data.
Margin of error: The amount added to and subtracted from the point estimate to create the confidence interval.
Confidence level: The probability that the true population parameter falls within the confidence interval.
Calculation of confidence interval:

The general formula for calculating a confidence interval for the mean is:

Confidence interval = Point estimate ± Margin of error
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
Ans:Bayes' Theorem: A Practical Example
Problem:

A rare disease affects 0.1% of the population. A diagnostic test for this disease is 99% accurate in detecting positive cases and 98% accurate in detecting negative cases. If a person tests positive for the disease, what is the probability that they actually have it?

Solution:

Let's define the events:

A: The person has the disease.
B: The person tests positive.
We are given:   

P(A) = 0.001 (prior probability of the disease)
P(B|A) = 0.99 (probability of testing positive given the disease)
P(B|A') = 0.02 (probability of testing positive given no disease, where A' is the complement of A)
We want to find P(A|B), the probability of having the disease given a positive test.

Using Bayes' Theorem:

P(A|B) = P(B|A) * P(A) / P(B)
First, we need to calculate P(B) using the law of total probability:

P(B) = P(B|A) * P(A) + P(B|A') * P(A')
      = 0.99 * 0.001 + 0.02 * 0.999
      = 0.02097
Now, we can apply Bayes Theorem:

P(A|B) = 0.99 * 0.001 / 0.02097
        ≈ 0.0472
```

```python
Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5. Interpret the results.
Ans:Calculating the 95% Confidence Interval
Given:

Sample mean (x̄) = 50
Sample standard deviation (s) = 5
Confidence level = 95%
Formula for a confidence interval:

CI = x̄ ± Z * (s / √n)
where:

CI is the confidence interval
Z is the z-score corresponding to the desired confidence level
s is the sample standard deviation
n is the sample size
Finding the Z-score:
For a 95% confidence level, the Z-score is approximately 1.96.   

Assuming a large enough sample size (n > 30), we can use the sample standard deviation as an estimate of the population standard deviation.   

Calculating the confidence interval:

CI = 50 ± 1.96 * (5 / √n)
```

```python
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
Ans:Margin of error in a confidence interval is the amount added to and subtracted from the point estimate (sample mean) to create the upper and lower bounds of the interval. It represents the uncertainty associated with the estimate.

Relationship between sample size and margin of error:

Inversely proportional: As the sample size increases, the margin of error decreases.
Intuitively: A larger sample size provides more information about the population, leading to a more precise estimate and a smaller margin of error.
Example:

Suppose you want to estimate the average height of adult males in a city. You take two random samples: one with a sample size of 100 and another with a sample size of 1000.

Smaller sample (n=100): The margin of error will be relatively large, meaning the estimate will be less precise.
Larger sample (n=1000): The margin of error will be smaller, indicating a more precise estimate.
```

```python
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.
Ans:Calculating the Z-Score
Given:

Data point (x) = 75
Population mean (μ) = 70
Population standard deviation (σ) = 5
Formula for z-score:

z = (x - μ) / σ
Substituting the values:

z = (75 - 70) / 5
   = 5 / 5
   = 1
Interpretation:

A z-score of 1 indicates that the data point (75) is 1 standard deviation above the population mean (70).
This means that the data point is relatively high compared to the overall population.
In a standard normal distribution, a z-score of 1 corresponds to approximately the 84th percentile. This means that 84% of the data points in the population are below 75.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
Ans:Hypothesis Testing for Weight Loss Drug Effectiveness
Given:

Sample size (n) = 50
Sample mean (x̄) = 6 pounds
Sample standard deviation (s) = 2.5 pounds
Confidence level = 95%
Hypothesis:

Null hypothesis (H₀): The drug has no effect on weight loss (μ = 0).
Alternative hypothesis (H₁): The drug is effective in promoting weight loss (μ > 0).
Test statistic (t-score):

t = (x̄ - μ) / (s / √n)
Substituting the values:

t = (6 - 0) / (2.5 / √50)
   ≈ 16.97
Degrees of freedom:

df = n - 1 = 50 - 1 = 49
Critical value:
Using a t-table with 49 degrees of freedom and a 95% confidence level, the critical value is approximately 1.677.
```

```python
Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.
Ans:Calculating the 95% Confidence Interval for a Proportion
Given:

Sample size (n) = 500
Sample proportion (p̂) = 0.65
Confidence level = 95%
Formula for the confidence interval of a proportion:

CI = p̂ ± Z * √(p̂ * (1 - p̂) / n)
where:

CI is the confidence interval
Z is the z-score corresponding to the desired confidence level   
p̂ is the sample proportion   
n is the sample size
Finding the Z-score:
For a 95% confidence level, the Z-score is approximately 1.96.

Substituting the values:

CI = 0.65 ± 1.96 * √(0.65 * (1 - 0.65) / 500)
   ≈ 0.65 ± 1.96 * 0.0215
   ≈ 0.65 ± 0.0421
Therefore, the 95% confidence interval is approximately (0.6079, 0.6921).
```
