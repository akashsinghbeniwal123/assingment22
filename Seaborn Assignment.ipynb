```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Ans:Seaborn is a powerful Python library built on top of Matplotlib, providing a high-level interface for creating attractive statistical visualizations. Here are five commonly used plot types in Seaborn and their applications:

Line Plots:

Use: Visualizing trends and relationships between two variables over time or a continuous range.
Example: Plotting stock prices over time, tracking changes in temperature, or analyzing sales data.
Scatter Plots:

Use: Showing the distribution of data points in two dimensions, often used to identify correlations or patterns.
Example: Visualizing the relationship between height and weight in a population, analyzing the correlation between income and education level.
Bar Plots:

Use: Comparing categorical data, representing values as bars.
Example: Comparing sales figures across different product categories, analyzing the distribution of age groups in a survey.
Histograms:

Use: Visualizing the distribution of a single numerical variable, showing the frequency of data points within certain intervals.
Example: Understanding the distribution of income levels in a country, analyzing the distribution of test scores in a class.
Box Plots:

Use: Comparing the distribution of multiple datasets, showing the median, quartiles, and potential outliers.
Example: Comparing the distribution of salaries across different industries, analyzing the variability of test scores between different schools.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.title("fMRI Signal Over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.boxplot(x='pclass', y='age', data=titanic, ax=axes[0])
axes[0].set_title('Age by Passenger Class')
sns.boxplot(x='pclass', y='fare', data=titanic, ax=axes[1])
axes[1].set_title('Fare by Passenger Class')
plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)
plt.title('Diamond Price Distribution by Cut')
plt.show()
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue='species')
plt.show()
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights_pivoted = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivoted, annot=True, fmt='d')
plt.title('Monthly Passenger Flights by Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
