```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.
Ans:alculating and Interpreting Pearson Correlation Coefficient
Understanding the Problem:

We have two variables: Study Time and Final Exam Score.
We need to calculate the Pearson correlation coefficient between these variables.
Pearson Correlation Coefficient (r):

Measures the strength and direction of the linear relationship between two variables.
Ranges from -1 to 1:
r = 1: Perfect positive correlation (variables increase or decrease together)
r = -1: Perfect negative correlation (one variable increases as the other decreases)
r = 0: No correlation
  
Formula:

r = (nΣxy - ΣxΣy) / sqrt((nΣx^2 - (Σx)^2)(nΣy^2 - (Σy)^2))
where:

n = number of data points
Σxy = sum of the product of corresponding values of x and y
Σx = sum of x values
Σy = sum of y values
Steps:

Collect data: Gather data on Study Time and Final Exam Score for a group of students.
Calculate the necessary sums: Σx, Σy, Σx^2, Σy^2, and Σxy.
Plug the values into the formula: Calculate the Pearson correlation coefficient using the formula.
Interpret the result:
If r is close to 1, there is a strong positive correlation between Study Time and Final Exam Score.
If r is close to -1, there is a strong negative correlation (which would be unusual in this case).
If r is close to 0, there is little or no correlation.
```

```python
Q2. Spearmans rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.
Ans:Calculating and Interpreting Spearman's Rank Correlation
Understanding the Problem:

We have two variables: Sleep Duration and Job Satisfaction.
Both variables are ordinal (measured on a scale).
We need to calculate the Spearmans rank correlation coefficient.
Spearmans Rank Correlation (ρ):

Measures the monotonic relationship between two variables.
Ranges from -1 to 1:
ρ = 1: Perfect positive monotonic relationship
ρ = -1: Perfect negative monotonic relationship
ρ = 0: No monotonic relationship
Formula:

ρ = 1 - 6Σd^2 / (n(n^2 - 1))
where:

d = difference in rank between corresponding pairs
n = number of data points
Steps:

Rank the data: Rank each variable separately from lowest to highest.
Calculate the difference in rank for each pair: d = Rank(x) - Rank(y).
Square the differences: d^2.
Sum the squared differences: Σd^2.
Plug the values into the formula: Calculate the Spearman rank correlation coefficient.
Interpret the result:
If ρ is close to 1, there is a strong positive monotonic relationship.
If ρ is close to -1, there is a strong negative monotonic relationship.
If ρ is close to 0, there is little or no monotonic relationship.
```
